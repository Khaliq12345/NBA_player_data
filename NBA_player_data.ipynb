```python
import pandas as pd
```

```python
l = []

alphabet = [
    'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o',
    'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z'
]
for alpha in alphabet:
    print('Scraping table', alpha)
    url = f'https://www.basketball-reference.com/players/{alpha}/'
    table = pd.read_html(url)
    l.append(table[0])

NBA_player = pd.concat([
    l[0], l[1], l[2], l[3], l[4], l[5], l[6], l[7], l[8], l[9], l[10], l[11],
    l[12], l[13], l[14], l[15], l[16], l[17], l[18], l[19], l[20], l[21],
    l[22], l[23], l[24], l[25]
])
NBA_player.to_csv('NBA_players_data.csv')
NBA_player
```
